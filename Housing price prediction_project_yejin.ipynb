## Model2: Partial Least Squares (PLS)

```r
# Method1: PLS Package (plsr() function) for Partial Least Squares Regression.
set.seed(123)
pls_model <- plsr(Median_House_Value ~ ., data = train.dat, validation = "CV")
summary(pls_model)

# Plot RMSEP vs Number of Components
plot(RMSEP(pls_model), legendpos = "topright")
```

### Interpretation:
### According to the validaiton plot, we may use 8 PLS component in the model

```r
# Method2: Caret Package (train() function) for automated cross-validation and model tuning.
set.seed(123)
ctrl <- trainControl(method = "repeatedcv", repeats = 5)

pls.fit <- train(
  Median_House_Value ~ .,        
  data = train.dat,              
  method = "pls",                
  preProc = c("center", "scale"), 
  tuneLength = 8,                
  trControl = ctrl               
)
summary(pls.fit)

# Predict on the test data
yhat_pls <- predict(pls.fit, newdata = test.dat)

# mse_pls1 <- round(mean((test.dat$Median_House_Value - yhat_pls)^2),3)
# mse_pls1
# Find the best number of components
best_component <- pls.fit$bestTune
print(best_component)

# Predict on the test data
yhat_pls <- predict(pls.fit, newdata = test.dat)

rmse_pls <- round(sqrt(mean((test.dat$Median_House_Value - yhat_pls)^2)),3)
rmse_pls

#calculate R^2
r2_pls <- round(cor(yhat_pls, test.dat$Median_House_Value)^2,3)
r2_pls
```

## Model3: Lasso Regression

```r
x_train <- model.matrix(Median_House_Value ~ ., data = train.dat)[, -1]  
y_train <- train.dat$Median_House_Value
x_test <- model.matrix(Median_House_Value ~ ., data = test.dat)[, -1]  
y_test <- test.dat$Median_House_Value

set.seed(123)
# Fit Lasso model using cross-validation
lasso_model <- cv.glmnet(x_train, y_train, alpha = 1, family = "gaussian")
# Plot the cross-validation results
plot(lasso_model)

# Get the best lambda that minimizes cross-validation error
best_lambda <- lasso_model$lambda.min
print(best_lambda)
# Get the coefficients for the Lasso model at the best lambda
lasso_coefficients <- coef(lasso_model, s = best_lambda)
print(lasso_coefficients)
```

### Lasso Regression Equation

$$
House\ Price = 20.8695 + 7.1368 \cdot Median\_Income + 1.1286 \cdot Median\_Age + \cdots \\
- 2.2438 \cdot Distance\_to\_SanJose + 2.4813 \cdot Distance\_to\_SanFrancisco
$$


```r
# Predict on the test data using the optimal lambda
yhat_lasso <- predict(lasso_model, s = best_lambda, newx = x_test)
# Calculate Mean Squared Error (MSE)
rmse_lasso <- round(sqrt(mean((y_test - yhat_lasso)^2)),3)
print(rmse_lasso)

# Calculate R^2
r2_lasso <- round(cor(yhat_lasso, y_test)^2,3)
print(r2_lasso)
```

```r
# Plot Actual vs Predicted Values for Lasso Regression
x <- 1:length(y_test)
plot(x, y_test, col = "red", type = "l", lwd = 2,
     main = "Testing Data: Actual vs Predicted Values (Lasso Regression)",
     xlab = "Index", ylab = "Median House Value")
lines(x, yhat_lasso, col = "blue", lwd = 2)
legend("topright", legend = c("Actual", "Predicted (Lasso)"),
       col = c("red", "blue"), lty = 1, lwd = 2)
grid()
```

# Model 5:  Neural network model

```r
set.seed(42)

#data preprocessing
#min-max normalization

scale.fun = function(x) {
  return ((x - min(x)) / (max(x) - min(x)))
}

miny = min(dat$Median_House_Value)
range.y = max(dat$Median_House_Value) - miny

#transformed data 

housing.dat = dat%>% mutate_all(scale.fun)

#split the data

index = createDataPartition(housing.dat$Median_House_Value, p = 0.8, list = FALSE)
train.dat = housing.dat[index,]
test.dat = housing.dat[-index,]

# build a nerual network on the training data
nn.fit = neuralnet(Median_House_Value ~ ., data = train.dat, hidden = c(2,3), linear.output = TRUE)
plot(nn.fit,rep="best")
```

```r
#make prediction    

set.seed(123)
#to make prediction using the testing data
yhat.0 = predict(nn.fit,test.dat)

#transform the result back to the original scale
yhat = yhat.0 * range.y + miny
test.dat$Median_House_Value = test.dat$Median_House_Value * range.y + miny


plot(yhat, test.dat$Median_House_Value,
     xlab = "Predicted by Neural network model", # X-axis label
     ylab = "Actual")    # Y-axis label

abline(0, 1, col = "red") # Add a diagonal line

#correlation 
```

```r

#RMSE on the testing data
rmse_nnt=round(sqrt(mean((yhat - test.dat$Median_House_Value)^2)),3)
rmse_nnt

#R^2 on the testing data
r2_nnt=round(cor(yhat, test.dat$Median_House_Value)^2,3)
r2_nnt
```

```r
library(kernlab)
set.seed(123)

# Train-Test Split
Index <- createDataPartition(sdat$Median_House_Value, p = 0.8, list = FALSE)
train.dat <- sdat[Index, ]
test.dat <- sdat[-Index, ]

#cross validation 
cv.control = trainControl(method = "cv", number = 5)

#create a tuning grid
#hyper parameter : C and sigma

hyper_grid = expand.grid(C=c(0.25,.5,1,1.5,2.0),sigma=c(.1,.15,.2))
#train to find the optimal tuning parameters

svm.fit= train(Median_House_Value~., data=train.dat, method = "svmRadial", 
            trControl = cv.control,
            trace = FALSE,
            tuneGrid = hyper_grid
            #PreProc = c("center","scale")
            #maxit = 500
)


#the best tuning parameter 
svm.fit$bestTune
```

# Model 8 : CART

```r
library(rpart)

index = createDataPartition(sdat$Median_House_Value, p = 0.8, list = FALSE)
train.dat = sdat[index,]
test.dat = sdat[-index,]
```

```r
#rpart : Recursive Partitioning and Regression Trees
#rpart is another name of CART in R
library(rpart)
#build a CART on the training data
fit.rpart = rpart(Median_House_Value ~., data = train.dat)

#To show the regression tree
plot(fit.rpart)
text(fit.rpart, digits = 3)

#Make a predict on the testing data
yhat = predict(fit.rpart, test.dat )
```

```r
set.seed(123)
# Control parameters
cntrl <- rpart.control(minsplit = 10, minbucket =3, maxdepth = 20)

# Train CART model with correct method
fit2 <- rpart(
  Median_House_Value ~ ., 
  data = train.dat, 
  method = "anova",  # Use "anova" for regression problems
  control = cntrl
)

# Plot the tree
#left = YES, right = NO
plot(fit2)
text(fit2, digits = 3)
```

```r
set.seed(123)
#Cross validation : 10 folder cross validation
ctrl = trainControl(method = "cv", number = 10)
#Tuning parameter: the complexity parameter (cp),
#which imposes a penalty to the tree for having two many splits.
#The default value is 0.01. The higher the cp, the smaller the tree.
#A too small value of cp leads to overfitting and a too large cp value will re
#An optimal cp value can be estimated by testing different cp values and using
#To create the searching grid
tgrid = expand.grid(
cp= c(0.000001, 0.0001, 0.0002, 0.001, 0.002)
)
# fit.rpart1 = train(Median_House_Value ~., data = train.dat, 
#                     method = "rpart", trControl = ctrl, tuneGrid =tgrid)
fit.rpart1 = train(Median_House_Value  ~., data = train.dat, 
                    method = "rpart", trControl = ctrl, tuneLength =10)
fit.rpart1
fit.rpart1$bestTune
```

```r
#Train a CART (rpart) by caret train function
set.seed(123)
#Cross validation : 10 folder cross validation
ctrl = trainControl(method = "cv", number = 10)
fit.rpart2 = train(Median_House_Value ~., data = train.dat,
                    method = "rpart", trControl = ctrl, tuneLength =20)
fit.rpart2
```

## Interpretation : 

#### Here’s what the decision tree shows: The most important factor is Distance_to_coast. Being closer to the coast leads to higher housing price. The second key factor is Median_Income, where areas with lower income usually have lower predicted values. Other factors like Latitude and Distance_to_SanJose also help refine the predictions. This tree helps us understand how these factors work together to make predictions in a simple and clear way."

```r
# Make predictions
yhat <- predict(fit2, newdata = test.dat)

# Plot Actual vs Predicted Values
x <- 1:nrow(test.dat)
plot(x, test.dat$Median_House_Value, col = "red", type = "l", lwd = 2,
     main = "Testing Data: Actual vs Predicted Values (CART Model)",
     xlab = "Index", ylab = "Median House Value")
lines(x, yhat, col = "blue", lwd = 2)
legend("topright", legend = c("Actual", "Predicted (CART)"),
       col = c("red", "blue"), lty = 1, lwd = 2)
grid()
```

```r
# Evaluate performance
rmse_cart <- round(sqrt(mean((yhat - test.dat$Median_House_Value)^2)),3)

# Calculate R^2
r2_cart <- round(cor(yhat, test.dat$Median_House_Value)^2,3)

print(paste("RMSE:", rmse_cart))
print(paste("R^2:", r2_cart))
```
